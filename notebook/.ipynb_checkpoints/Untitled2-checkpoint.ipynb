```python
import re
def read_and_extract(
    files, 
    pattern,
    **kwargs
    ):
    """
    Parameters:
        files: List of paths to dataframes to read
        pattern: Regex pattern with group for extracting cluster id.
            "Kmeans(\d+).csv"
            "ShapeCluster(\d+)"
        kwargs: Arguments passed to :func:`pandas.read_csv`
    Returns:
        Dictionary mapping cluster ID to its dataframe
    """
    files.sort()
    clusters = [re.findall(pattern,f)[0] for f in files]
    cluster2df = {c:pd.read_csv(f, **kwargs) for c, f in zip(clusters, files)}
    return cluster2df
comps = ['csf', 'pbmc']

# read kmeans and shape clusters

shape_files = {c: glob.glob("../data/tcr_dynamics/{}ShapeCluster*.csv".format(c.upper()))
    for c in comps}
shape_files = {c: glob.glob("../../paper_upn109/data/tcr/{}ShapeCluster*.csv".format(c.upper()))
    for c in comps}

shapes = {c:read_and_extract(shape_files[c], "ShapeCluster(\d+).csv") 
    for c in comps}
```

```python
def logic2label(logic, groups):
    return '∩'.join([g for l,g in zip(logic, groups) if l=='1'])
```

```python
pre =  tumor.loc[tumor['Treatment']=='Pre']['TRB'].tolist()
post  =  tumor.loc[tumor['Treatment']=='Post']['TRB'].tolist()
```

```python
### Get intersections with CAR and Tumor ####
groups =  ['Shape', 'Pre', 'Post']
dfs = []
for c in comps:
    for i, df in shapes[c].items():
        print(c, i, len(df)) # check shape size
        shape_name = "{}:{}".format(c.upper(), i)
        shape = df['TRB'].tolist()
        subsets = get_subsets([shape, pre, post])
        sizes = get_subset_sizes(subsets)
        # get subsets containing shape
        shape_logic = [k for k in sizes if k[0]=='1']
        shape_sizes = {k:sizes[k] for k in shape_logic}
        df_subsets = pd.DataFrame(shape_sizes, index=[shape_name]).rename(
            columns=lambda x: logic2label(x, groups))
        dfs.append(df_subsets)
```

```python
df_all = pd.concat(dfs)
subset_names = df_all.columns
df_all['sum'] = df_all.sum(axis=1) # check shape size
```

```python
def get_trb2rank(df_top, count_col):
    # remove non-top clonotypes
    df_top = df_top[df_top.index.str.startswith('TRA')]
    # rank by cell counts
    df_top = df_top.sort_values(count_col, ascending=False)
    # extract TRB
    df_top['TRB'] = df_top.index.to_series().str.split('TRB:', expand=True)[1]
    df_top = df_top.reset_index()
    display(df_top)
    
    trb2rank = {}
    for i, row in df_top.iterrows():
        trb2rank[row['TRB']] = i+1
    return trb2rank
```

```python
top_pre = get_trb2rank(df_top_pre, "Pre")
top_pre
```

```python
top_post = get_trb2rank(df_top_post, "Post")
top_post
```

```python
pb_noise = df_mapped.loc[df_mapped['Sample:ShapeCluster']=='PB:Noise', 'TcRb'].values
```

```python
timepoints = pb_t
timeseries = pb
ncols = 4
nrows = math.ceil(len(timepoints) / ncols)
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols*3, nrows*3), sharex=True)
fig.suptitle('L-infininity')
for t, ax in zip(timepoints, axes.reshape(-1)):
    x = get_nnz(timeseries[t]).sort_values()
    rank = x.rank(ascending=True, method='first', pct=True)
    mode = x.mode().values[0]
    top = ~x.index.isin(pb_noise)
#     fig, ax = plt.subplots() 
    ax.scatter(rank, x, c=top, s=1, cmap='bwr')
    ax.set_ylim(-0.0001, ax.get_ylim()[1])
    min_f = x.loc[rank.loc[top].idxmin()]
    filtering = '{}/{} min={:.4f}'.format(top.sum(), len(x), min_f)
    ax.set_title(t+'\n'+filtering)
    plt.tight_layout()
```

```python
timepoints = csf_t
timeseries = csf
ncols = 4
nrows = math.ceil(len(timepoints) / ncols)
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols*3, nrows*3), sharex=True)
fig.suptitle('Filter Mode')
signals = set()
for t, ax in zip(timepoints, axes.reshape(-1)):
    x = get_nnz(timeseries[t]).sort_values()
    rank = x.rank(ascending=True, method='first', pct=True)
    mode = x.mode().values[0]
    top = x > mode
#     fig, ax = plt.subplots() 
    ax.scatter(rank, x, c=top, s=1, cmap='bwr')
    ax.set_ylim(-0.0001, ax.get_ylim()[1])
    signals.update(x.loc[top].index.values)
    min_f = x.loc[rank.loc[top].idxmin()]
    filtering = '{}/{} cutoff={:.4f}'.format(top.sum(), len(x), min_f)
    ax.set_title(t+'\n'+filtering)
    plt.tight_layout()
```
