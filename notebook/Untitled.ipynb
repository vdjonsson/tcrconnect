```python
import pandas as pd
```

```python
df = pd.read_csv('../data/gliph/blood-csf-tumor/gliph_output_blood-csf-tumor_general.csv')
```

```python
trb_col='TcRb'
```

```python
inds_trb = df[trb_col] == 'CSARLAGDYNEQFF'
```

```python
patterns = df.loc[inds_trb, 'pattern'].unique()
```

```python
inds_pattern = df['pattern'].isin(patterns)
```

```python
df_pattern = df.loc[inds_pattern, ['pattern', 'TcRb','TcRa', 'Sample', 'Freq']]
df_pattern.to_csv('CSARLAGDYNEQFF_specificity_groups.csv', index=False)
```
